# What is the BEST FEATURE?

In SKLEARN:

- **CART (default)**: Uses
  - Gini Index (Impurity)
    - Gini = 1 - sum(pow(p1, 2))
    - Can only answer YES or NO (binary tree branches)
- **ID3**:
  - Entropy
    - Entropy = E(S) = sum(-p1 * log2(p2))
  - Information Gain = Entropy(Before) - Entropy(After)
  - Can only answer multiple answers
