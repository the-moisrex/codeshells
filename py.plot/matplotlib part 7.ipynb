```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
import numpy as np

# plt.style.use("seaborn")
```

```python
x = np.random.randint(1, 10, 50)
y = np.random.randint(1, 10, 50)
x, y
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y,
           s=300)
plt.show()
```

```python
plt.scatter(x, y,
           s=30,
           c='green')
plt.show()
```

```python
plt.scatter(x, y,
           s=30,
           c='green',
           marker='X')
plt.show()
```

```python
plt.scatter(x, y,
           s=30,
           c='green',
           edgecolor='black',
           linewidth=1,
           alpha=0.3)
plt.show()
```

```python
colors = np.arange(len(x))
plt.scatter(x, y,
           s=30,
           c=colors,
           edgecolor='black',
           linewidth=1,
           alpha=0.3)
plt.show()
```

```python
colors = np.arange(len(x))
plt.scatter(x, y,
           s=30,
           c=colors,
           cmap='Greens',
           edgecolor='black',
           linewidth=1,
           alpha=0.5)
plt.show()
```

```python
colors = np.arange(len(x))
plt.scatter(x, y,
           s=30,
           c=colors,
           cmap='Greens',
           edgecolor='black',
           linewidth=1,
           alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction Level')

plt.show()
```

```python
colors = np.arange(len(x))
sizes = np.random.randint(100, 500, len(x))

plt.scatter(x, y,
           s=30,
           c=colors,
           sizes=sizes,
           cmap='Greens',
           edgecolor='black',
           linewidth=1,
           alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction Level')

plt.show()
```

```python
colors = np.arange(len(x))
sizes = np.random.randint(100, 500, len(x))

plt.scatter(x, y,
           s=30,
           c=colors,
           sizes=sizes,
           cmap='summer',
           edgecolor='black',
           linewidth=1,
           alpha=0.5)

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Satisfaction Level')

plt.show()
```
