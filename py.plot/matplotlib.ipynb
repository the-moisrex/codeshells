```python
plt.plot(dev_x, dev_y)
plt.plot(py_dev_x, py_dev_y)
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend(['All Developers', 'Python Developers']) # by order
plt.show()
```

```python
plt.plot(dev_x, dev_y, label='All Devs')
plt.plot(py_dev_x, py_dev_y, label='Py Devs')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend() # it will use the labels from up there
plt.show()
```

```python

# fmt = '[marker][line][color]'
plt.plot(dev_x, dev_y, 'k--', label='All Devs')       # k = black
plt.plot(py_dev_x, py_dev_y, 'r.-', label='Py Devs')  # ,- = line
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend() # it will use the labels from up there
plt.show()
```

```python
# Easier to use arguments than to remember all those styles in the format string

plt.plot(dev_x, dev_y,
         color='#09a', linestyle='--', marker='.', linewidth=1, label='All Devs')
plt.plot(py_dev_x, py_dev_y,
         color='r', linestyle='-.', marker='o', linewidth=3, label='Py Devs')
plt.plot(py_dev_x, js_dev_y,
         color='#aa3', linestyle='-', marker='.', label='JS Devs')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend()
plt.tight_layout()   # Doesn't seem to have an effect
plt.show()
```
