# Matplotlib Tutorial (Part 5): Filling Area on Line Plots

```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
import numpy as np

plt.style.use("fivethirtyeight")
```

```python
data = pd.read_csv('dev_salaries_5.csv')

ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.2)

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = np.median(dev_salaries)

plt.fill_between(ages, py_salaries, overall_median, alpha=0.2)

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.2)

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = np.median(dev_salaries)
plt.fill_between(ages, py_salaries, overall_median,
                 alpha=0.2,
                 where=(py_salaries > overall_median)
                )

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = np.median(dev_salaries)
plt.fill_between(ages, py_salaries, overall_median,
                 alpha=0.2,
                 interpolate=True,
                 where=(py_salaries > overall_median)
                )

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#333',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = np.median(dev_salaries)
plt.fill_between(ages, py_salaries, overall_median,
                 alpha=0.2,
                 interpolate=True,
                 color='green',
                 where=(py_salaries > overall_median)
                )
plt.fill_between(ages, py_salaries, overall_median,
                 alpha=0.2,
                 interpolate=True,
                 where=(py_salaries <= overall_median)
                )

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries,
         color='#333',
         linewidth=1,
         label='All Devs')

plt.plot(ages, py_salaries,
         label='Python',
         linewidth=1
        )

plt.fill_between(ages, py_salaries, dev_salaries,
                 alpha=0.2,
                 interpolate=True,
                 color='green',
                 where=(py_salaries > dev_salaries)
                )
plt.fill_between(ages, py_salaries, dev_salaries,
                 alpha=0.2,
                 color='red',
                 interpolate=True,
                 where=(py_salaries <= dev_salaries)
                )

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

```python
plt.plot(ages, dev_salaries,
         color='#333',
         linewidth=1,
         label='All Devs')

plt.plot(ages, py_salaries,
         label='Python',
         linewidth=1
        )

plt.fill_between(ages, py_salaries, dev_salaries,
                 alpha=0.2,
                 interpolate=True,
                 color='green',
                 where=(py_salaries > dev_salaries),
                 label='Above Avg'
                )
plt.fill_between(ages, py_salaries, dev_salaries,
                 alpha=0.2,
                 color='red',
                 interpolate=True,
                 where=(py_salaries <= dev_salaries),
                 label='Bellow Avg'
                )

plt.legend()
plt.title('Median Salary (USD) by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```
