```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
dates = [
    datetime(2023, 1, 1),
    datetime(2023, 1, 12),
    datetime(2023, 1, 21),
    datetime(2023, 1, 13),
    datetime(2023, 1, 16),
    datetime(2023, 1, 17),
    datetime(2023, 1, 13),
    datetime(2023, 1, 12),
    datetime(2023, 1, 17),
    datetime(2023, 1, 18),
    datetime(2023, 1, 1),
    datetime(2023, 1, 6),
    datetime(2023, 1, 5),
    datetime(2023, 1, 2),
    datetime(2023, 1, 5),
    datetime(2023, 1, 7),
    datetime(2023, 1, 2),
    datetime(2023, 1, 1),
]
```

```python
y = np.arange(len(dates))

plt.plot_date(dates, y)
plt.show()
```

```python
plt.plot_date(dates, y,
             linestyle='solid')
plt.show()
```

```python
plt.plot_date(dates, y,
             linestyle='solid')
# "get current figure"
plt.gcf().autofmt_xdate()
plt.show()
```

```python
dates = np.sort(dates)
plt.plot_date(dates, y,
             linestyle='solid')
# "get current figure"
plt.gcf().autofmt_xdate()
plt.show()
```

```python
date_formater = mpl_dates.DateFormatter('%b, %d %Y')

plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()

# "Get Current Axis"
plt.gca().xaxis.set_major_formatter(date_formater)
```

```python
data = pd.read_csv('dates.csv')
price_date = data['Date']
price_close = data['Close']


plt.plot_date(price_date,
              price_close,
              linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
data = pd.read_csv('dates.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_valuesalues('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']


plt.plot_date(price_date,
              price_close,
              linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```
