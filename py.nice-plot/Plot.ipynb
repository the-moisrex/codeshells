```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

fm = matplotlib.font_manager
fm._get_fontconfig_fonts.cache_clear()
```

```python
white = '#fff'
gray = '#999'
black = '#111'
blue = '#16a2c6'
font = 'Roboto'
```

```python
uber = pd.DataFrame({
    'City': [
        'New York City',
        'Washington D.C.',
        'Chicago',
        'Los Angles',
        'Dalas'
    ],
    'Uber': [142, 96, 77, 62, 181],
    'Personal Car': [218, 130, 116, 89, 65]
})

uber
```

```python
fig, ax = plt.subplots(figsize=(3, 6), dpi=100)

(uber
 .set_index('City')
 .T
 .plot(ax=ax, 
       linewidth=1.7,
       color=[blue if x == 'Dalas' else gray for x in uber.City],
       legend=None,
       marker='o',
       zorder=5)
)
```

```python
# Tick location
ax.set_xticks([0, 1],
              ['Uber', 'Personal Car'],
              font=font,
              size=12,
              weight=600)
ax.set_yticks([])

fig
```

```python
# Remove ticks
ax.xaxis.set_tick_params(bottom=False)
ax.yaxis.set_tick_params(left=False)

fig
```

```python
# Spine color
for side in 'left,top,right,bottom'.split(','):
    ax.spines[side].set_visible(False)

fig
```

```python
# Vertical Lines
ax.vlines([0, 1],
          [50, 50],
          [230, 230],
          color=gray,
          linewidth=0.2)

fig
```

```python
# Labels
for idx, value in uber.set_index('City').Uber.items():
    if idx == 'Dallas':
        color = blue
    else:
        color = gray
    ax.text(x=0.02,
            y=value,
            s=f'${value}',
            color=color,
            va='center',
            ha='right',
            font=font,
            weight=600)

fig
```

```python
for idx, value in uber.set_index('City')['Personal Car'].items():
    if idx == 'Dallas':
        color = blue
    else:
        color = gray
    ax.text(x=1.02,
            y=value,
            s=f'{idx}, ${value}',
            color=color,
            va='center',
            ha='left',
            font=font,
            weight=600)
```

```python
# Add in title and subtitle
ax.text(x=0,
        y=0.93,
        s='Weekly cost of commute by city',
        transform=fig.transFigure,
        ha='left',
        font=font,
        fontsize=10,
        weight=1000)
```

```python
ax.text(x=0,
        y=0.9,
        s=r"The estimated costs are based on a twice daily 10.4 mile commute",
        font=font,
        fontsize=11,
        weight=300,
        color=black)

fig
```
