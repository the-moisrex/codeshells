```python
import matplotlib.pyplot as plt
import numpy as np

NUM_VALUES = 1000

def generate_x_values(num_values):
    x = list(range(num_values))
    return x

def generate_y_values(x):
    y = [2 * v for v in x]
    return y

def linear_regression(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    num = 0
    den = 0
    for i in range(len(x)):
        num += (x[i] - x_mean) * (y[i] - y_mean)
        den += (x[i] - x_mean) ** 2
    slope = num / den
    intercept = y_mean - slope * x_mean
    return slope, intercept

x = generate_x_values(NUM_VALUES)
y = generate_y_values(x)

# Perform linear regression on the data
m, b = linear_regression(x, y)

# Generate data points for the regression line
x_reg = np.array([min(x), max(x)])
y_reg = m * x_reg + b

# Plot the data as a scatter plot
plt.scatter(x, y)
plt.plot(x_reg, y_reg, color='red')
plt.title("x vs. y (Linear Regression)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values for x and y
x = np.random.rand(1000)
y = x * 2 + np.random.rand(1000)

# Linear regression function
def linear_regression(x, y):
    n = len(x) # number of observations
    x_mean = sum(x) / n # mean of x variable
    y_mean = sum(y) / n # mean of y variable

    numerator = 0
    denominator = 0

    for i in range(n):
        numerator += (x[i] - x_mean) * (y[i] - y_mean)
        denominator += (x[i] - x_mean) ** 2

    slope = numerator / denominator
    intercept = y_mean - slope * x_mean
    
    return slope, intercept

# Calculate slope and intercept coefficients
slope, intercept = linear_regression(x, y)
print("Slope: ", slope)
print("Intercept: ", intercept)

# Predict function
def predict(x_test, slope, intercept):
    y_pred = slope * x_test + intercept
    return y_pred

# Example predictions
x_test = np.array([0.1, 0.3, 0.6, 0.9])
y_pred = predict(x_test, slope, intercept)
print("Predictions:", y_pred)

# Create scatter plot of x and y
plt.scatter(x, y, s=10)
plt.xlabel('X')
plt.ylabel('Y')

# Add predictions to the plot
plt.scatter(x_test, y_pred, color='red')

plt.show()
```
